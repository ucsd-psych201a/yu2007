```python
import os
import random
```

```python
with open("text.txt", encoding="utf-8") as f:
    text = f.read().splitlines()
text = [line for line in text if not line.isupper()]
text
```

```python
random.seed(3407)
word_list = []
for i in range(54):
    sample = random.sample(text, k=2)
    pseudoword = sample[0] + sample[1]
    word_list.append(pseudoword)
word_list
```

```python
with open("word.txt", "w") as w:
    for word in word_list:
        w.write(word+"\n")
```
